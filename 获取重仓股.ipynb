```python
import akshare as ak
# 一季报
df1 = ak.fund_report_stock_cninfo(date="20230331")
# 二季报/半年报
df2 = ak.fund_report_stock_cninfo(date="20230630")
# 三季报
df3 = ak.fund_report_stock_cninfo(date="20230930")
# 年报/四季报
df4 = ak.fund_report_stock_cninfo(date="20231231")
merged_df = df1.merge(df2, on=['股票代码'], suffixes=('_df1', '_df2'))

# 计算差值
merged_df['序号变化'] = merged_df['序号_df2'] - merged_df['序号_df1']
merged_df['基金覆盖家数变化'] = merged_df['基金覆盖家数_df2'] - merged_df['基金覆盖家数_df1']
merged_df['持股总数变化'] = merged_df['持股总数_df2'] - merged_df['持股总数_df1']
merged_df['持股总市值变化'] = merged_df['持股总市值_df2'] - merged_df['持股总市值_df1']

# 只保留需要的 diff 列
diff_df = merged_df[['股票代码', '股票简称_df1', '序号变化', '基金覆盖家数变化', '持股总数变化', '持股总市值变化']]
diff_df = diff_df.sort_values(by='基金覆盖家数变化', ascending=False).reset_index(drop=True)
diff_df
```
