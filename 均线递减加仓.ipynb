```python
import pybroker as pyb
from pybroker import ExecContext,PriceType
import talib as ta

def getIndicator():
    sma_10 = pyb.indicator('sma_10', lambda data: ta.SMA(data.close, timeperiod=10))
    sma_20 = pyb.indicator('sma_20', lambda data: ta.SMA(data.close, timeperiod=20))
    sma_30 = pyb.indicator('sma_30', lambda data: ta.SMA(data.close, timeperiod=30))
    sma_40 = pyb.indicator('sma_40', lambda data: ta.SMA(data.close, timeperiod=40))
    sma_50 = pyb.indicator('sma_50', lambda data: ta.SMA(data.close, timeperiod=50))
    sma_60 = pyb.indicator('sma_60', lambda data: ta.SMA(data.close, timeperiod=60))
    sma_70 = pyb.indicator('sma_70', lambda data: ta.SMA(data.close, timeperiod=70))
    sma_80 = pyb.indicator('sma_80', lambda data: ta.SMA(data.close, timeperiod=80))
    sma_90 = pyb.indicator('sma_90', lambda data: ta.SMA(data.close, timeperiod=90))
    sma_250 = pyb.indicator('sma_250', lambda data: ta.SMA(data.close, timeperiod=250))
    sma_120 = pyb.indicator('sma_120', lambda data: ta.SMA(data.close, timeperiod=120))
    sma_100 = pyb.indicator('sma_100', lambda data: ta.SMA(data.close, timeperiod=100))
    return [sma_10,sma_20,sma_30,sma_40,sma_50,sma_60,sma_70,sma_80,sma_90,sma_120,sma_250,sma_100]
# 拥有所有的context
# def before_run():
#     return
# 只有一个context
def run(ctx: ExecContext):
    print("进入rotate")
    # 判断当前k线到了哪个价位
    dsignal_10 = (ctx.close[-1] > ctx.indicator('sma_10')[-1]) & (ctx.close[-2] <= ctx.indicator('sma_10')[-2])
    dsignal_20 = (ctx.close[-1] > ctx.indicator('sma_20')[-1]) & (ctx.close[-2] <= ctx.indicator('sma_20')[-2])
    dsignal_30 = (ctx.close[-1] > ctx.indicator('sma_30')[-1]) & (ctx.close[-2] <= ctx.indicator('sma_30')[-2])
    dsignal_40 = (ctx.close[-1] > ctx.indicator('sma_40')[-1]) & (ctx.close[-2] <= ctx.indicator('sma_40')[-2])
    dsignal_50 = (ctx.close[-1] > ctx.indicator('sma_50')[-1]) & (ctx.close[-2] <= ctx.indicator('sma_50')[-2])
    dsignal_60 = (ctx.close[-1] > ctx.indicator('sma_60')[-1]) & (ctx.close[-2] <= ctx.indicator('sma_60')[-2])
    dsignal_70 = (ctx.close[-1] > ctx.indicator('sma_70')[-1]) & (ctx.close[-2] <= ctx.indicator('sma_70')[-2])
    dsignal_80 = (ctx.close[-1] > ctx.indicator('sma_80')[-1]) & (ctx.close[-2] <= ctx.indicator('sma_80')[-2])
    dsignal_90 = (ctx.close[-1] > ctx.indicator('sma_90')[-1]) & (ctx.close[-2] <= ctx.indicator('sma_90')[-2])
    print(f"dsignal_10:{dsignal_10},dsignal_20:{dsignal_20},dsignal_30:{dsignal_30},dsignal_40:{dsignal_40},")
    if ctx.close[-1]<ctx.indicator('sma_100')[-1]:
        print("风控")
        return
    if ctx.long_pos():
        return
    if dsignal_10:
        print(f"{ctx.dt}:买入")
        print(f"当日价位大于sma10:{ctx.close[-1]}>{ctx.indicator('sma_10')[-1]}")
        print(f"昨日价位小于等于昨日sma10:{ctx.close[-2]}>{ctx.indicator('sma_10')[-2]}")
        print(f"ctx.close[-1] :{ctx.close[-1]}")
        print(f"ctx.close[-2]:{ctx.close[-2]}")
        print(f"ctx.indicator('sma_20')[-1]:{ctx.indicator('sma_20')[-1]}")
        print(f"ctx.indicator('sma_20')[-2]:{ctx.indicator('sma_20')[-2]}")
        ctx.buy_shares = ctx.calc_target_shares(0.2)
        ctx.stop_loss_pct=5
        ctx.stop_loss_exit_price=PriceType.CLOSE
        # 追踪止损/追踪止损价位
        ctx.stop_trailing_pct = 10
        ctx.stop_trailing_exit_price = PriceType.CLOSE
        # 止盈
        ctx.stop_profit_pct=5
    if dsignal_20:
        ctx.buy_shares = ctx.calc_target_shares(0.4)
        ctx.stop_loss_pct=5
        ctx.stop_loss_exit_price=PriceType.CLOSE
        # 追踪止损/追踪止损价位
        ctx.stop_trailing_pct = 10
        ctx.stop_trailing_exit_price = PriceType.CLOSE
        # 止盈
        ctx.stop_profit_pct=10
    if dsignal_30:
        ctx.buy_shares = ctx.calc_target_shares(0.6)
        ctx.stop_loss_pct=5
        ctx.stop_loss_exit_price=PriceType.CLOSE
        # 追踪止损/追踪止损价位
        ctx.stop_trailing_pct = 10
        ctx.stop_trailing_exit_price = PriceType.CLOSE
        # 止盈
        ctx.stop_profit_pct=20
    if dsignal_40:
        ctx.buy_shares = ctx.calc_target_shares(0.8)
        ctx.stop_loss_pct=5
        ctx.stop_loss_exit_price=PriceType.CLOSE
        # 追踪止损/追踪止损价位
        ctx.stop_trailing_pct = 10
        ctx.stop_trailing_exit_price = PriceType.CLOSE
        # 止盈
        ctx.stop_profit_pct=30
    if dsignal_50:
        ctx.buy_shares = ctx.calc_target_shares(1)
        ctx.stop_loss_pct=5
        ctx.stop_loss_exit_price=PriceType.CLOSE
        # 追踪止损/追踪止损价位
        ctx.stop_trailing_pct = 10
        ctx.stop_trailing_exit_price = PriceType.CLOSE
        # 止盈
        ctx.stop_profit_pct=40
    # if dsignal_60:
    #     ctx.buy_shares = ctx.calc_target_shares(0.6)
    #     ctx.stop_loss_pct=5
    #     ctx.stop_loss_exit_price=PriceType.CLOSE
    #     # 追踪止损/追踪止损价位
    #     ctx.stop_trailing_pct = 10
    #     ctx.stop_trailing_exit_price = PriceType.CLOSE
    #     # 止盈
    #     ctx.stop_profit_pct=30
    # if dsignal_70:
    #     ctx.buy_shares = ctx.calc_target_shares(0.7)
    #     ctx.stop_loss_pct=5
    #     ctx.stop_loss_exit_price=PriceType.CLOSE
    #     # 追踪止损/追踪止损价位
    #     ctx.stop_trailing_pct = 10
    #     ctx.stop_trailing_exit_price = PriceType.CLOSE
    #     # 止盈
    #     ctx.stop_profit_pct=35
    # if dsignal_80:
    #     ctx.buy_shares = ctx.calc_target_shares(0.8)
    #     ctx.stop_loss_pct=5
    #     ctx.stop_loss_exit_price=PriceType.CLOSE
    #     # 追踪止损/追踪止损价位
    #     ctx.stop_trailing_pct = 10
    #     ctx.stop_trailing_exit_price = PriceType.CLOSE
    #     # 止盈
    #     ctx.stop_profit_pct=40
    # if dsignal_90:
    #     ctx.buy_shares = ctx.calc_target_shares(0.9)
    #     ctx.stop_loss_pct=5
    #     ctx.stop_loss_exit_price=PriceType.CLOSE
    #     # 追踪止损/追踪止损价位
    #     ctx.stop_trailing_pct = 10
    #     ctx.stop_trailing_exit_price = PriceType.CLOSE
    #     # 止盈
    #     ctx.stop_profit_pct=45
```

```python
import pybroker as pyb
from pybroker import Strategy,StrategyConfig
from data.data_resouce import AKShare_ALL
config = StrategyConfig(initial_cash=500000,max_long_positions=1,exit_on_last_bar=True)
ak=AKShare_ALL()
stock_code="600809"
df=ak.query(symbols=[stock_code],start_date="20170801", end_date="20241112",adjust="hfq")
strategy = Strategy(df, start_date="20170801", end_date="20241112", config=config)
# strategy.set_before_exec(before)
# ,"000902","000904","000016","399313","399300","399316","399967","399378","399402"
strategy.add_execution(run, [stock_code], indicators=getIndicator())
result = strategy.backtest(warmup=120)
print('======基准收益率==========================')
start_open = df[df['date'] == "20170801"]["open"].values[0]
end_open = df[df['date'] == "20240801"]["open"].values[0]
last_open = df['open'].iloc[-1]
benchmark_return = (end_open - start_open) / start_open * 100
print(f"基准收益率{benchmark_return}")
```
