```python
from data.data_resouce import AKShare_ALL
import pybroker as pyb
from pybroker import ExecContext, Strategy, StrategyConfig, YFinance
import talib as ta
import akshare as ak
from auto.strategy_chart import create_strategy_charts
roc_20 = pyb.indicator(
    'roc_20', lambda data: ta.ROC(data.close, timeperiod=20))
pyb.enable_data_source_cache('my1_strategy')
config = StrategyConfig(max_long_positions=2)
pyb.param('target_size', 1 / config.max_long_positions)
pyb.param('rank_threshold', 5)

def rank(ctxs: dict[str, ExecContext]):
    scores = {
        symbol: ctx.roc_20[-1]
        for symbol, ctx in ctxs.items()
    }
    print(f"scores:{scores}")
    sorted_scores = sorted(
        scores.items(),
        key=lambda score: score[1],
        reverse=True
    )
    print(f"sorted_scores:{sorted_scores}")
    threshold = pyb.param('rank_threshold')
    top_scores = sorted_scores[:threshold]
    top_symbols = [score[0] for score in top_scores]
    pyb.param('top_symbols', top_symbols)

def rotate(ctx: ExecContext):
    if ctx.long_pos():
        if ctx.symbol not in pyb.param('top_symbols'):
            ctx.sell_all_shares()
    else:
        target_size = pyb.param('target_size')
        ctx.buy_shares = ctx.calc_target_shares(target_size)
        ctx.score = ctx.roc_20[-1]
import akshare as ak
aa=ak.index_stock_cons_csindex(symbol="000050")

akshare = AKShare_ALL()
df = akshare.query(symbols=aa[""].values[0:10], start_date="20170801", end_date="20190801",adjust="hfq")
df['roc_20'] = ta.ROC(df.close, timeperiod=20)
pyb.register_columns('roc_20')

strategy = Strategy(df, start_date="20170801", end_date="20190801", config=config)
strategy.set_before_exec(rank)
strategy.add_execution(rotate, aa[""].values[0:10])
result = strategy.backtest(warmup=20)
```
